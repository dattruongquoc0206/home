# BÃ i 2

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

m=100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)

X_new = np.linspace(-3, 3, 100).reshape(100, 1)
poly_features = PolynomialFeatures(degree=1, include_bias=False)
X_poly = poly_features.fit_transform(X)
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
poly_features = PolynomialFeatures(degree=1, include_bias=False)
X_poly_new = poly_features.fit_transform(X_new)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
y_poly_pred = lin_reg.predict(X_poly_new)

figure, axis = plt.subplots(1,3, figsize=(20, 25))
axis[0].plot(X_poly, y, "b.")
axis[0].plot(X_poly_new, y_poly_pred, "r-")
axis[0].set_title("degree 1")

X_new = np.linspace(-3, 3, 100).reshape(100, 1)
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly_new = poly_features.fit_transform(X_new)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
y_poly_pred = lin_reg.predict(X_poly_new)

axis[1].plot(X, y, "b.")
axis[1].plot(X_new, y_poly_pred, "r-")
axis[1].set_title("degree 2")


X_new = np.linspace(-3, 3, 100).reshape(100, 1)
poly_features = PolynomialFeatures(degree=30, include_bias=False)
X_poly = poly_features.fit_transform(X)
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
poly_features = PolynomialFeatures(degree=30, include_bias=False)
X_poly_new = poly_features.fit_transform(X_new)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
y_poly_pred = lin_reg.predict(X_poly_new)

axis[2].plot(X, y, "b.")
axis[2].plot(X_new, y_poly_pred, "r-")
axis[2].set_title("degree 30")

plt.show()

```

# BÃ i 3


```python
import numpy as np
import matplotlib.pyplot as plt

x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

X_b = np.c_[np.ones((100, 1)), x]
alpha = 0.02
I = np.identity(2)
theta_best = np.linalg.inv((X_b.T.dot(X_b))+(alpha*I)).dot(X_b.T).dot(y)
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_pred = X_new_b.dot(theta_best)

plt.plot(x, y, "b.")
plt.plot(X_new, y_pred, "r-")
plt.axis([0,2,0,15])
plt.suptitle("Ridge Regression")

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

figure, axis = plt.subplots(2,2, constrained_layout=True)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

#linear without sklearn
X_b = np.c_[np.ones((100, 1)), x]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_pred = X_new_b.dot(theta_best)
axis[0, 0].plot(x, y, "b.")
axis[0, 0].plot(X_new, y_pred, "r-")
axis[0, 0].set(xlim=(0,2), ylim=(0, 15))
axis[0, 0].set_title("linear without sklearn")

#Ridge without sklearn
X_b = np.c_[np.ones((100, 1)), x]
alpha = 0.02
I = np.identity(2)
theta_best = np.linalg.inv((X_b.T.dot(X_b))+(alpha*I)).dot(X_b.T).dot(y)
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_pred = X_new_b.dot(theta_best)
axis[0, 1].plot(x, y, "b.")
axis[0, 1].plot(X_new, y_pred, "r-")
axis[0, 1].set(xlim=(0,2), ylim=(0, 15))
axis[0, 1].set_title("ridge without sklearn")

#linear with sklearn
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
y_sklearn_pred = lin_reg.predict(X_new)
axis[1, 0].plot(x, y, "b.")
axis[1, 0].plot(X_new, y_pred, "r-")
axis[1, 0].set(xlim=(0,2), ylim=(0, 15))
axis[1, 0].set_title("linear with sklearn")

#Ridge with sklearn
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_sklearn_pred = lin_reg.predict(X_new)
axis[1, 1].plot(x, y, "b.")
axis[1, 1].plot(X_new, y_pred, "r-")
axis[1, 1].set(xlim=(0,2), ylim=(0, 15))
axis[1, 1].set_title("ridge with sklearn")


plt.show()
```
